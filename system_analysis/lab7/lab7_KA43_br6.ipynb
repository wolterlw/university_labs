```python
def plot_graph(table, save=True):
    G = nx.DiGraph(table.values)
    #plot >0 and <=0 weights 
    elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.0]
    esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 0.0]
    
    pos = nx.spring_layout(G)
    plt.figure(figsize=(12, 12),frameon=False)
    nx.draw_networkx_nodes(G, pos, node_size=750,
                          node_color='k')
    
    nx.draw_networkx_edges(G, pos, edgelist=elarge,
                           edge_color='r', alpha=0.5,
                           width=2)
    
    nx.draw_networkx_edges(G, pos, edgelist=esmall,
                           width=2, alpha=0.5, edge_color='b', style='dashed')

    nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif', font_color='white')

    plt.axis('off')
    if save:
        plt.savefig('graph.png', dpi=100)
    plt.show()
    
def is_stable_by_value(eigenvalues):
    eig_max = np.apply_along_axis(lambda x: abs(x), 0, eigenvalues).max()
    is_one = (eigenvalues == 1+0j).any()
    return is_one or (eig_max < 1)

def is_stable_by_perturbation(eigenvalues):
    eig_max = np.apply_along_axis(lambda x: abs(x), 0, eigenvalues).max()
    return eig_max <= 1

def is_stable_by_numerical(eigenvalues):
    eig_max = np.apply_along_axis(lambda x: abs(x), 0, eigenvalues).max()
    return eig_max < 1

def get_edges(cycle, delete_last=True):
    cycle = np.array(cycle)
    cycle_shifted = np.roll(cycle, -1)
    edges = [(element_1, element_2) for element_1, element_2 in zip(cycle, cycle_shifted)]
    if delete_last:
        del edges[-1]
    return edges

def is_even(edges, weights):
    result = 1
    for edge in edges:
        result *= weights[edge]
    return result

def is_stable_by_structure(table):
    matrix = table.values
    
    G = nx.DiGraph(matrix)
    weights = {}
    for (u, v, d) in G.edges(data=True):
        weights[(u, v)] = d['weight']
    
    gen = nx.simple_cycles(G)
    
    even_cycles = []
    n_even_cycles = 0
    n_odd_cycles = 0
    while True:
        try:
            cycle = next(gen)
        except StopIteration:
            break
        res = even_cycle(get_edges(cycle), weights)
        if res > 0:
            even_cycles.append(cycle)
            n_even_cycles += 1
        else:
            n_odd_cycles += 1
    shuffle(even_cycles)
    return even_cycles, n_even_cycles, n_odd_cycles

def get_most_frequent_vertex(even_cycles, table):
    lst = [[(c[i], c[i+1]) for i in range(len(c)-1)] for c in even_cycles] 
    vertices = []
    for c in lst:
        for v in c:
            vertices.append(v)
    most_frequent = max(set(vertices), key=vertices.count)
    return (most_frequent[0]+1,most_frequent[1]+1)
```
